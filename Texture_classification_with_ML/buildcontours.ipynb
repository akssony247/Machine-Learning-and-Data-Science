```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from skimage import measure
```

```python
im = cv2.imread("C:/Users/aksso/.spyder-py3/PulseTest/4frame608.jpg",cv2.IMREAD_GRAYSCALE)
a = np.asarray(im)

contours = measure.find_contours(a, 100, fully_connected='low',positive_orientation='low')
fig, ax = plt.subplots()
ax.imshow(a, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
